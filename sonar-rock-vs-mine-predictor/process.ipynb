```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
sonar_data = pd.read_csv("data/Copy of sonar data.csv")
sonar_data.head()
```

```python
sonar_data.shape
```

```python
sonar_data.iloc[:,60].value_counts()
```

```python
sonar_data_values = sonar_data.iloc[:,:60]
sonar_data_labels = sonar_data.iloc[:,-1]

sonar_data_values
```

```python
sonar_data_labels
```

```python
values_train, values_test, labels_train, labels_test = train_test_split(sonar_data_values, sonar_data_labels, test_size=0.1, stratify=sonar_data_labels, random_state=1)
print("Values : " , sonar_data_values.shape, values_train.shape, values_test.shape)
print("Labels : ", sonar_data_labels.shape, values_train.shape, values_test.shape)

print("Values Train : ", values_train.shape)
print("Labels Train : ", labels_train.shape)
print("Values Test : ", values_test.shape)
print("Labels Test : ", labels_test.shape)

```

```python
model = LogisticRegression()
model.fit(values_train, labels_train)
```

```python
print(values_train)
print(labels_train)
```

```python
values_train_prediction = model.predict(values_train)
training_data_accuracy = accuracy_score(values_train_prediction, labels_train)
```

```python
print("Accuracy : " , training_data_accuracy)
```

```python
values_test_prediction = model.predict(values_test)
test_data_accuracy = accuracy_score(values_test_prediction, labels_test)
```

```python
print("Accuracy : " , test_data_accuracy)
```
